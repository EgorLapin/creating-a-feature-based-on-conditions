```python
import pandas as pd
import numpy as np
import sklearn
# Загрузка
diabetes = sklearn.datasets.load_diabetes()
# Преобразование
table = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
# Условие: больше среднего индекса массы тела
condition = table['bmi'] > table['bmi'].mean()
# Создание признака. 1 => выполнено условие, 0 иначе
table['bmi_above_mean'] = np.where(condition, 1, 0)

print("Среднее значение bmi:", table['bmi'].mean(), "Первые 6 строк DataFrame с новым признаком:")
print(table.head(6))
print("Распределение нового признака:")
print(table['bmi_above_mean'].value_counts())
```
